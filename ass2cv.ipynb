```python
#Assignment CSPB441
import matplotlib.pyplot as plt
import numpy as np

# Define the vector components
x1, y1 = 5,2  # Vector a components
x2, y2 = 1,4 # Vector b components

# Define the vectors a and b
a = np.array([x1, y1])
b = np.array([x2, y2])

# 1. Drawing the Vectors on a 2D Coordinate System
plt.figure()
plt.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1, color='r', label='a = <x1, y1>')
plt.quiver(0, 0, b[0], b[1], angles='xy', scale_units='xy', scale=1, color='b', label='b = <x2, y2>')
plt.xlim(-1, max(x1, x2) + 1)
plt.ylim(-1, max(y1, y2) + 1)
plt.xlabel('X')
plt.ylabel('Y')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.grid(True)
plt.show()

# 2. Finding the Difference Between the Vectors
difference = a - b
print("Difference between the vectors:", difference)

# 3. Finding the Sum of the Vectors
sum_vectors = a + b
print("Sum of the vectors:", sum_vectors)

# 4. Finding the Angles alpha and theta
# 1. Find the angle α of vector a relative to the x-axis
alpha = np.arctan2(a[1], a[0])  # Angle α of vector a
alpha_deg = np.degrees(alpha)   # Convert from radians to degrees
print(f"Angle α (alpha) of vector a: {alpha_deg:.2f} degrees")

# 2. Find the angle θ between vectors a and b
dot_product = np.dot(a, b)  # Dot product of vectors a and b
norm_a = np.linalg.norm(a)   # Norm (magnitude) of vector a
norm_b = np.linalg.norm(b)   # Norm (magnitude) of vector b

cos_theta = dot_product / (norm_a * norm_b)
theta = np.arccos(cos_theta)  # Angle θ between vectors a and b
theta_deg = np.degrees(theta) # Convert from radians to degrees
print(f"Angle θ (theta) between vectors a and b: {theta_deg:.2f} degrees")
# 5. Calculating the Euclidean Norms
norm_a = np.linalg.norm(a)
norm_b = np.linalg.norm(b)
print(f"Euclidean norm of vector a: {norm_a:.2f}")
print(f"Euclidean norm of vector b: {norm_b:.2f}")

# 6. Dot Product Calculation
dot_product = np.dot(a, b)
print(f"Dot product of a and b: {dot_product}")

# 7. Parallelogram and Dot Product using Cosine
dot_product_cos = norm_a * norm_b * cos_theta
print(f"Dot product using cosine formula: {dot_product_cos:.2f}")

# Draw parallelogram
plt.figure()
plt.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1, color='r', label='a = <x1, y1>')
# Create dotted line effect by plotting multiple small dots

# Create dotted line effect with smaller dots

# Define the points for the parallelogram
points = np.array([[0, 0], a, a + b, b, [0, 0]])

# Draw the parallelogram with dotted lines
plt.plot(points[:, 0], points[:, 1], 'k:', label='Parallelogram')  # 'k:' means black dotted line

plt.title(f'Parallelogram and Dot Product\nDot Product: {dot_product:.2f}, Angle θ: {theta_deg:.2f}°')

plt.quiver(0, 0, b[0], b[1], angles='xy', scale_units='xy', scale=1, color='b', label='b = <x2, y2>')
plt.xlim(-1, max(a[0] + b[0], b[0] + a[0]) + 1)
plt.ylim(-1, max(a[1] + b[1], b[1] + a[1]) + 1)
plt.xlabel('X')
plt.ylabel('Y')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.grid(True)
plt.show()

# 8. Cross Product Consideration for 3D
a_3d = np.array([x1, y1, 0])
b_3d = np.array([x2, y2, 0])
cross_product = np.cross(a_3d, b_3d)
print("Cross product a × b:", cross_product)

#9. Cross Product using Sine and Visualization

a = np.array([4, 2, 0])
b = np.array([3, 5, 0])
cross_product = np.array([0, 0, 14])  # Cross product result
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the vectors
ax.quiver(0, 0, 0, a[0], a[1], a[2], color='r', label='a = [5, 2, 0]')
ax.quiver(0, 0, 0, b[0], b[1], b[2], color='b', label='b = [1, 4, 0]')
ax.quiver(0, 0, 0, cross_product[0], cross_product[1], cross_product[2], color='g', label='a × b')

# Setting the limits
ax.set_xlim([0, 10])
ax.set_ylim([0, 10])
ax.set_zlim([0, 20])

# Labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Visualization of Cross Product')

# Adding legend
ax.legend()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Given vector V in 2D and convert it to homogeneous coordinates
V = np.array([11, 2, 1])

# Given scaling matrix S
S = np.array([
    [4, 0, 0],
    [0, 3, 0],
    [0, 0, 1]
])

# Given translation matrix T
T = np.array([
    [1, 0, 4],
    [0, 1, 3],
    [0, 0, 1]
])

# Rotation matrix for 50 degrees (convert degrees to radians)
theta = np.radians(145)
R = np.array([
    [np.cos(theta), -np.sin(theta), 0],
    [np.sin(theta), np.cos(theta), 0],
    [0, 0, 1]
])

# 1. Scaled vector V' = S * V
V_scaled = S @ V
print("Scaled Vector V' = S * V:", V_scaled[:2])

# 2. Rotated vector V' = R * V
V_rotated = R @ V
print("Rotated Vector V' = R * V:", V_rotated[:2])

# 3. Scaled and rotated vector V' = R * S * V
V_scaled_rotated = R @ V_scaled
print("Scaled and Rotated Vector V' = R * S * V:", V_scaled_rotated[:2])

# 4. Scaled, rotated, and translated vector V' = T * R * S * V
V_scaled_rotated_translated = T @ V_scaled_rotated
print("Scaled, Rotated, and Translated Vector V' = T * R * S * V:", V_scaled_rotated_translated[:2])

# 5. Homogeneous transformation matrix H = T * R * S
H = T @ R @ S
print("Homogeneous Transformation Matrix H = T * R * S:\n", H)

# 6. Transformed vector V' = H * V
V_transformed = H @ V
print("Transformed Vector V' = H * V:", V_transformed[:2])

# Plotting the original and transformed vectors

# Original vector
V_original = V[:2]

# Transformed vectors in 2D
V_scaled_2d = V_scaled[:2]
V_rotated_2d = V_rotated[:2]
V_scaled_rotated_2d = V_scaled_rotated[:2]
V_scaled_rotated_translated_2d = V_scaled_rotated_translated[:2]

# Create the plot
plt.figure(figsize=(8, 8))
plt.grid(True)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)

# Plot original vector
plt.quiver(0, 0, V_original[0], V_original[1], angles='xy', scale_units='xy', scale=1, color='blue', label='Original V')

# Plot scaled vector
plt.quiver(0, 0, V_scaled_2d[0], V_scaled_2d[1], angles='xy', scale_units='xy', scale=1, color='green', label='Scaled V')

# Plot rotated vector
plt.quiver(0, 0, V_rotated_2d[0], V_rotated_2d[1], angles='xy', scale_units='xy', scale=1, color='red', label='Rotated V')

# Plot scaled and rotated vector
plt.quiver(0, 0, V_scaled_rotated_2d[0], V_scaled_rotated_2d[1], angles='xy', scale_units='xy', scale=1, color='purple', label='Scaled & Rotated V')

# Plot scaled, rotated, and translated vector
plt.quiver(0, 0, V_scaled_rotated_translated_2d[0], V_scaled_rotated_translated_2d[1], angles='xy', scale_units='xy', scale=1, color='orange', label='Scaled, Rotated & Translated V')

# Set limits and labels
plt.xlim(-40, 40)
plt.ylim(-40, 40)
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.title('Transformations of Vector V')

# Show the plot
plt.show()
```
